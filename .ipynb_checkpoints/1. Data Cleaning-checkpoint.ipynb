## 1. Data Cleaning

Los datos, su limpieza, interpretación y procesamiento son la base de cualquier algoritmo de Machine Learning, por lo tanto, debemos explicar qué tipo de datos tenemos, cómo los tratamos y que características sacamos de los mismos.

Toda esta información esta incluida en el research de: Yang Bao, Bin Ke, Bin Li, Julia Yu, and Jie Zhang (2020). Detecting Accounting Fraud in Publicly Traded U.S. Firms Using a Machine Learning Approach. Journal of Accounting Research, 58 (1): 199-235. Donde se ha logrado combinar toda esta información y añadir algunas variables financieras calculada que los expertos piensan que serian vitales para la detección del fraude. 

El archivo "uscecchini28.csv" del año 2000 es nuestro conjunto de datos final que contiene las etiquetas de fraude y las variables de características. 
El nombre de variable dependiente " misstate " (1 denota fraude y 0 denota no fraude). 
El resto de variables se clasifican en 2 tipos de variables: 
- Variables financieras: dch_wc, ch_rsst, dch_rec, dch_inv, soft_assets, ch_cs, ch_cm, ch_roa, issue, bm, dpi, reoa, EBIT, ch_fcf y new_p_aaer.
- Variables financieras calculadas: dch_wc, ch_rsst, dch_rec, dch_inv, soft_assets, ch_cs, ch_cm, ch_roa, issue, bm, dpi, reoa, EBIT y ch_fcf.

La diferencia entre las dos es que las variables financieras calculadas son resultado de cálculos en base a las variables financieras. 

Para asegurarnos que nuestro modelo pueda predecir bien sin necesidad de la ayuda de los expertos y científicos del research, debe ser capaz de ofrecer óptimos resultados sin incluir las variables financieras calculadas, por tanto, debemos eliminarlas.

```python
# Visualización de las variables:
df.hist(figsize =(30,30))
plt.show()
```

```python
# Vamos a dividir el dataset por años. 
years = list(df.fyear.unique())
fraud_per_year = []

for year in years:
    fraud_per_year.append((df[df['fyear'] == year].groupby('gvkey').agg({'misstate':'sum'}).sum()/len
                           (df[df['fyear'] == year].groupby('gvkey').agg({'misstate':'sum'}))*100))
```

```python
df.info()
# la mayoria de las columnas tienen 146045 registros, algunas no. verificamos la existencia de "Nan"
```

"count" nos muestra que, sí tenemos datos faltantes, ya que las columnas no coinciden en el número total de observaciones. Esto nos indica que, puede haber Nan o datos vacíos.

```python
# Número de nan por columna
len(df) - df.count()
```

```python
# Sustituir los “Nan” con la media de cada columna.    
df['sich'] = df['sich'].fillna(df['sich'].median())
df['p_aaer'] = df['p_aaer'].fillna(df['p_aaer'].median())
df['new_p_aaer'] = df['new_p_aaer'].fillna(df['new_p_aaer'].median())
df['dch_wc'] = df['dch_wc'].fillna(df['dch_wc'].median())
df['ch_rsst'] = df['ch_rsst'].fillna(df['ch_rsst'].median())
df['dch_rec'] = df['dch_rec'].fillna(df['dch_rec'].median())
df['dch_inv'] = df['dch_inv'].fillna(df['dch_inv'].median())
df['soft_assets'] = df['soft_assets'].fillna(df['soft_assets'].median())
df['ch_cs'] = df['ch_cs'].fillna(df['ch_cs'].median())
df['ch_cm'] = df['ch_cm'].fillna(df['ch_cm'].median())
df['ch_roa'] = df['ch_roa'].fillna(df['ch_roa'].median())
df['bm'] = df['bm'].fillna(df['bm'].median())
df['dpi'] = df['dpi'].fillna(df['dpi'].median())
df['reoa'] = df['reoa'].fillna(df['reoa'].median())
df['EBIT'] = df['EBIT'].fillna(df['EBIT'].median())
df['ch_fcf'] = df['ch_fcf'].fillna(df['ch_fcf'].median())

print("Actualizamos el Número de registros en nuestro Dataframe: ",len(df))
```

```python
# Número de nan por columna: Verificamos que ya no tenemos Nan
df.isna().sum()
```

```python
# Eliminamos los valores duplicados 
df = df.drop_duplicates(keep = 'first')
print("Actualizamos el Número de registros en nuestro Dataframe: ",len(df))
```

```python
# Número de valores únicos en cada columna:Podemos los tipos de datos que tenemos
for i in range(0, len(df.columns)):
    print(df.columns[i], ":", df.iloc[:,i].nunique())
```

```python
# Seleccionamos los datos del año 2000 para nuestro análisis y lo guardamos como Csv    
df = df[(df.fyear >= 2000) & (df.fyear <= 2000)]
df.reset_index(drop=True, inplace=True)
df.reset_index(drop=True, inplace=True)
print(df.shape)
```

```python
# Lo guardamos 
df.to_csv('clean_uscecchini28.csv')
```

### Informe:

El dataset estaba en formato “Csv”. Venia mezclados los años, así que, hemos tenido que organizarlos por años; de esta forma será más fácil seleccionar el año que necesitamos y facilita su análisis. 

En cuanto a las anomalías, había algunas columnas que tenían un número importante de “Nan”; se ha optado por sustituirlos por la media de cada columna ya que la estructura de datos que presenta es bastante equilibrada siempre que miremos su media; y la distancia entre la media y el máximo de cada columna no es mucho. 

No se han borrado ninguna columna en esta fase debido a que, de momento, desconocemos la importancia que pueden tener alguna de las variables. Es mejor no precipitarse.
