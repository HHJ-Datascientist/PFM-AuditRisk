```python
# Importamos la librer√≠as necesarias

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%pylab
%matplotlib inline

# Ponemos la semilla
np.random.seed(0)
```

### Algortimos individuales: LR, RF y DT

```python
# Importamos el dataset 
individual = pd.read_csv('Individual.csv')
```

```python
individual = individual.drop(["Model Name.1", "Model Parameter"], axis = 1)
```

```python
individual = individual.sort_values('AUC',ascending=False)
```

```python
sns.set(style="whitegrid", context="notebook")
sns.set()
plt.rcParams['figure.figsize']=10,5
sns.barplot(x='AUC', y='Model Name', data=individual, label="AUC", color='b')
plt.title("Mejor cuanto m√°s cerca de 100")
plt.show();
```

```python
individual = individual.sort_values('F1',ascending=False)
```

```python
sns.set(style="whitegrid", context="notebook")
sns.set()
plt.rcParams['figure.figsize']=10,5
sns.barplot(x='F1', y='Model Name', data=individual, label="F1", color='b')
plt.title("Mejor cuanto m√°s cerca de 100")
plt.show();
```
